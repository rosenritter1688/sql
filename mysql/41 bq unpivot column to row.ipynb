### SYNTAX
```
SELECT * FROM `table_wanna_unpivot`
UNPIVOT ( column_name_for_original_row_values FOR column_name_for_original_column_name IN (
                                                                sport,	
                                                                watching_sport,	
                                                                travel,	mobility,	
                                                                reading,	
                                                                music,	
                                                                movie,	
                                                                computer,	
                                                                game,	
                                                                gamble))
ORDER BY
        points DESC
```
