#### get all indiviual values and union all

```python
sql_query = """
SELECT
                product,
                Q1 as sales,
                'q1' as quater
FROM
                product

UNION ALL

SELECT 
                product,
                Q2 as sales,
                'q2' as quater
FROM 
                product

UNION ALL


SELECT 
                product,
                Q3 as sales,      # GET COLUMN Q3 VALUES AND NAME COLUMN AS sales
                'q3' as quater  # give a string value 'q3' AND NAME COLUMN NAME AS QUATER
FROM
                product

UNION ALL

SELECT
                product,
                Q4 as sales,
                'q4' as quater
FROM
                product

"""

df = pd.read_sql_query(
    sql_query,
    mydb)


df
```
