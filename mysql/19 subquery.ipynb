```python
sql_query = """
SELECT *
FROM (
    SELECT *
    FROM order_report_1
    UNION ALL
        SELECT *
        FROM  order_report_2
) AS order_report;
"""


df = pd.read_sql_query(
    sql_query,
    mydb)


df
```

```python
sql_query = """
SELECT taste, sum(amount)
FROM (
    SELECT *
    FROM order_report_1
    UNION ALL
        SELECT *
        FROM  order_report_2
) AS order_report    
GROUP BY taste;
"""


df = pd.read_sql_query(
    sql_query,
    mydb)


df
```

```python
sql_query = """
SELECT order_report.date_ymd, 
                sum(amount) AS amount_day
-- subquery "order_report"
-- UNION ALL table order_report_1 and order_report_2
FROM (     
    SELECT *
    FROM order_report_1
    UNION ALL
        SELECT *
        FROM order_report_2
) AS order_report
GROUP BY date_ymd
ORDER BY amount_day DESC
;
"""


df = pd.read_sql_query(
    sql_query,
    mydb)


df
```
