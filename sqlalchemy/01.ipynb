```python
from sqlalchemy import create_engine, ForeignKey, Column,String,Integer,CHAR,VARCHAR 
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker

Base = declarative_base()

# 
class Person(Base):
    __tablename__ = "people"  
    ssn = Column("ssn", Integer, primary_key=True) 
    firstname = Column("firstname", VARCHAR(25) ) # use String for sqllite
    lastname = Column("lastname", VARCHAR(25))
    gender = Column("gender", CHAR)
    age = Column("age", Integer)

    def __init__(self, ssn, first, last, gender, age):
        self.ssn = ssn
        self.firstname = first
        self.lastname = last
        self.gender = gender
        self.age = age

    def __repro__(self):
        return f"{self.ssn} {self.firstname} {self.lastname} {self.gender} {self.age}"



engine = create_engine('mysql+mysqlconnector:// root:19821205@127.0.0.1/sql_fr_ai?charset=utf8', echo=True)
Base.metadata.create_all(bind=engine)

Session = sessionmaker(bind=engine)
session = Session()

person = Person(12311,"Mile","Smith","m",35)
session.add(person)
session.commit()

p1= Person(12312,"Anna","Blue","f", 40)
p2= Person(12313,"Bob","Blue","f", 40)
```
